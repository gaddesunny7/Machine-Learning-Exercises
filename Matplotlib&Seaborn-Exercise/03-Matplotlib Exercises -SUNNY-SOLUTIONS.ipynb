```python
import pandas as pd
import numpy as np 
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.9,0.9])
axes.plot(x,y,'b')
axes.set_xlabel('Set X Label')
axes.set_xlabel('Set Y Label')
axes.set_title('Set Title')
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0,0,1,1])
axes2 = fig1.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.plot(x,y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
fig,axes = plt.subplots(nrows =1,ncols=2)
```

```python
fig,axe1s = plt.subplots(nrows =1,ncols=2,figsize=(12,3))
axe1s[0].plot(x,y,'r',lw=3,ls='--')
axe1s[1].plot(x,z,'b',lw=3,ls='-')
```
