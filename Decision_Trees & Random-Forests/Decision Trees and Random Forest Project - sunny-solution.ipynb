```python
cat_feats = ['purpose']
final_data = pd.get_dummies(data,columns=cat_feats,drop_first=True)
```

```python
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(final_data.drop('not.fully.paid',axis=1),final_data[['not.fully.paid']],test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,Y_train)
```

```python
ypredict = dtree.predict(X_test)
```

```python
dtree.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_test,ypredict))
```

```python
print(confusion_matrix(Y_test,ypredict))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand = RandomForestClassifier(n_estimators=100)
rand.fit(X_train,Y_train)
```

```python
ypredict = rand.predict(X_test)
```

```python
rand.score(X_test,Y_test)
```
